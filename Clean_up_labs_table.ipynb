```python
### For Oxygen saturation, 95% is considered typical. <=92 medical issue. Take the min

df_lab_top_tests_OxSatu = df_lab_top_tests[df_lab_top_tests\
                                                       ['test_name']=='Oxygen saturation (SpO2).pulse oximetry'].dropna()

# Take min
df_lab_top_tests_OxSatu_min = df_lab_top_tests_OxSatu.groupby('patientid')[['test_result']]\
.min().reset_index().sort_values(by='patientid')

df_lab_top_tests_OxSatu_min.columns = ['patientid', 'OxSatu_min']

```

## Aggregate all tables into one, row as patientid, column as tests

```python
Labs_table_list = [df_lab_top_tests_OxSatu_min, \
                   df_lab_top_tests_Glucose_max, \
                   df_lab_top_tests_Creatinine_max, \
                   df_lab_top_tests_WBC_max, \
                   df_lab_top_tests_RBC_max, \
                   df_lab_top_tests_PLT_max, \
                   df_lab_top_tests_CO2_max, \
                   df_lab_top_tests_Neutro_max, \
                   df_lab_top_tests_Lympho_max, \
                   df_lab_top_tests_Mono_max, \
                   df_lab_top_tests_Eosin_max, \
                   df_lab_top_tests_Baso_max]
```

```python
df_lab_pid_for_left_merge = df_lab_top_tests[['patientid']].drop_duplicates()
df_lab_pid_for_left_merge['status'] ='PT_with_LAB'
```

```python
df_lab_pid_for_left_merge.shape
```

```python
df_lab_pid_for_left_merge.head()
```

```python
## left merge

df = df_lab_pid_for_left_merge
for obs_ind_table in Labs_table_list:
    print("shape of the table to be merged", obs_ind_table.shape)
    print(obs_ind_table.head())
    df = df.merge(obs_ind_table, on='patientid', how='left').sort_values(by='patientid')
    print("shape of intermediate table", df.shape)
    print(df.head())
    print("========")
```

```python
df_lab_pivot = df.drop(['status'], axis=1)
```
